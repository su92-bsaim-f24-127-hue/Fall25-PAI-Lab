```python
x_c = int(input("Enter capacity of Jug X:"))
y_c = int(input("Enter capacity of Jug Y:"))
goal = int(input("Enter goal amount:"))
start = (0, 0)
visited = set()
stack = [(start, [])]
def dfs(x_c, y_c):

    def rule1(x, y):   
        return (x_c, y)
    def rule2(x, y):   
        return (x, y_c)
    def rule3(x, y):   
        return (0, y)
    def rule4(x, y):   
        return (x, 0)
    def rule5(x, y):  
        transfer = min(x, y_c - y)
        return (x - transfer, y + transfer)
    def rule6(x, y):   
        transfer = min(y, x_c - x)
        return (x + transfer, y - transfer)
    def rule7(x, y):   
        total = x + y
        if total <= y_c:
            return (0, total)
        return (x, y)
    def rule8(x, y):   
        total = x + y
        if total <= x_c:
            return (total, 0)
        return (x, y)
    return [
        ("Fill X", rule1),
        ("Fill Y", rule2),
        ("Empty X", rule3),
        ("Empty Y", rule4),
        ("Pour X to Y", rule5),
        ("Pour Y to X", rule6),
        ("Transfer All X to Y", rule7),
        ("Transfer All Y to X", rule8),
    ]

rules = dfs(x_c, y_c)
found = False
while stack:
    (x, y), path = stack.pop()
    if x == goal or y == goal:
        print("Goal Found")
        for step in path:
            print(step)
        print(f"Final State: X = {x}L , Y = {y}L")
        found = True
        break
    if (x, y) in visited:
        continue
    visited.add((x, y))
    for name, rule in rules:
        new_state = rule(x, y)
        if new_state not in visited:
            stack.append((new_state, path + [f"{name} = {new_state}"]))

if not found:
    print("Goal not found")
```
