```python
def solve_n_queens(n):

    coloums = set()
    diag1 = set()  
    diag2 = set()   

    board = [["-" for _ in range(n)] for _ in range(n)]
    solutions = []

    def backtrack(row):
        if row == n:
            solutions.append([" ".join(r) for r in board])
            return

        for col in range(n):

            if col in coloums or (row-col) in diag1 or (row+col) in diag2:
                continue

            board[row][col] = "Q"
            coloums.add(col)
            diag1.add(row-col)
            diag2.add(row+col)

            backtrack(row + 1)

            board[row][col] = "-"
            coloums.remove(col)
            diag1.remove(row-col)
            diag2.remove(row+col)

    backtrack(0)

    print(f"No of Solutions: {len(solutions)}")

    for i, sol in enumerate(solutions, 1):
        print(f"Solution {i}:")
        for row in sol:
            print(row)
        print()


n = int(input("Enter number of queens: "))
solve_n_queens(n)
```
